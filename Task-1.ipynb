```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
iris_data = load_iris()
```

```python
features = pd.DataFrame(iris_data.data,columns = iris_data.feature_names)
features
```

```python
targets = iris_data.target
```

```python
def checkAccuracyBasedOnSpliting(features, targets, train_size=0.7):
    
    # split the data
    X_train, X_test, y_train, y_test = train_test_split(features, 
                                                        targets, 
                                                        random_state=30, 
                                                        train_size=train_size)
    # model
    my_model = DecisionTreeClassifier()
    
    #fit trains the model on train data
    my_model.fit(X_train, y_train)
    
    #Testing means pass on the unseen hidden test set to the model and ask it to do predictions
    y_pred = my_model.predict(X_test)

    # Compare the predictions - y_pred with the actuals - y_test
    print(train_size,"\t\t",accuracy_score(y_pred, y_test))
```

```python
print("Split size", "\t", "Accuracy")
print("(On train data)\n")
print("*"*38)
for i in [0.1,0.15,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,0.99]:
    checkAccuracyBasedOnSpliting(features, targets, train_size=float(i))
```
